## Building a RAG Application – ChromaDB Vector Store


## STEP 1 — Install and Import Libraries

```python
# Install necessary libraries: chromadb for vector store, sentence-transformers for embeddings, and langchain for potential future use.
# Import chromadb client and SentenceTransformer for creating and managing the vector database and generating embeddings.
!pip install chromadb sentence-transformers langchain --quiet

import chromadb
from sentence_transformers import SentenceTransformer
```

## STEP 2 — Initialize ChromaDB

```python
# Initialize the ChromaDB client.
# Create a new collection named 'rag_collection' in ChromaDB to store our documents and their embeddings.
client = chromadb.Client()
collection = client.get_or_create_collection(name="rag_collection")
```

## STEP 3 — Load Embedding Model

```python
# Load the 'all-MiniLM-L6-v2' Sentence Transformer model.
# This model will be used to convert text into dense vector embeddings.
model = SentenceTransformer('all-MiniLM-L6-v2')
```

## STEP 4 — Prepare Dataset

```python
# Define a list of sample documents that will be stored in the ChromaDB.
# Generate unique IDs for each document based on their index.
documents = [
    "Artificial Intelligence is the simulation of human intelligence.",
    "Machine Learning is a subset of AI.",
    "Deep Learning uses neural networks.",
    "Natural Language Processing deals with text data.",
    "RAG combines retrieval and generation.",
    "AI is used in healthcare and finance.",
    "Python is popular in machine learning.",
    "Data science involves statistics and programming.",
    "Neural networks are inspired by the brain.",
    "Transformers are used in NLP models."
]

ids = [f"doc{i}" for i in range(1, len(documents)+1)]
```

## STEP 5 — Convert Text to Embeddings

```python
# Encode the defined documents into vector embeddings using the loaded Sentence Transformer model.
embeddings = model.encode(documents)
```

## STEP 6 — Store in ChromaDB

```python
# Add the documents, their generated embeddings, and their corresponding IDs to the ChromaDB collection.
collection.add(
    documents=documents,
    embeddings=embeddings,
    ids=ids
)
```

## STEP 7 — Similarity Search

```python
# Define a query string.
# Encode the query into a vector embedding.
# Perform a similarity search in the ChromaDB collection to retrieve the top 2 most similar documents to the query.
# Print the raw results of the similarity search.
query = "What is AI?"

query_embedding = model.encode([query])

results = collection.query(
    query_embeddings=query_embedding,
    n_results=2
)

print(results)
```

## STEP 8 — Display Results

```python
# Iterate through the retrieved documents from the search results and print each one.
for doc in results['documents'][0]:
    print(doc)
```

## STEP 9 — Retrieval Function

```python
# Define a retrieval function that takes a query, embeds it, and searches the ChromaDB for the top 2 relevant documents.
# Demonstrate the use of the `retrieve` function with a new query.
def retrieve(query):
    query_embedding = model.encode([query])
    results = collection.query(query_embeddings=query_embedding, n_results=2)
    return results['documents'][0]

print(retrieve("Explain machine learning"))
```
